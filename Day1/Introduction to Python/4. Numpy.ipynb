## Numpy For image Processing 

```python
import cv2
import matplotlib.pyplot as plt
```

### Image grayscaling 

grayscale image = (R channel + g channel + b channel)/3

```python
# read image
image = cv2.imread("love.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image2, cmap='gray')
plt.show()
```

```python
ktp_ori = cv2.imread("ktp.jpg")
ktp = cv2.cvtColor(ktp_ori, cv2.COLOR_BGR2RGB)
ktp = cv2.cvtColor(ktp, cv2.COLOR_BGR2GRAY)
print (ktp.shape)
plt.imshow(ktp, cmap='gray')
plt.show()
```

```python
th3 = cv2.adaptiveThreshold(ktp, 191, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
plt.imshow(th3, cmap='gray')
plt.show()
```

```python
ret3,th4 = cv2.threshold(ktp,0,  255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(th4, cmap='gray')
plt.show()
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(th4, kernel, iterations = 4)

plt.imshow(dilation, cmap='gray')
plt.show()
```

```python

sobelx = cv2.Sobel(dilation,cv2.CV_64F,1,0,ksize=5)
sobelx = np.uint8(sobelx)
plt.imshow(sobelx)
plt.show()
```

```python
kernel = np.ones((9,9),np.uint8)
closing = cv2.morphologyEx(th4, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray')
plt.show()
```

```python
contours,hierarchy = cv2.findContours(dilation, 1, 2)
cnt = contours[3]
```

```python
hull = cv2.convexHull(cnt)
print (hull)
```
