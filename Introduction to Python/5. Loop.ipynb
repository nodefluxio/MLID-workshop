## While Loop

```python
error = 50.0
while error > 1 :
    error = error / 4
    print(error)
```

```python
error = 50.0
while error > 1: # BAHAYA ALWAYS TRUE INFINITY LOOP 
    # error = error / 4
    print(error)
```

## For Loop

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
print(keluarga)
```

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
print(keluarga[0])
print(keluarga[1])
print(keluarga[2])
print(keluarga[3])
```

```python
keluarga = [1.73, 1.68, 1.71, 1.89]
for tinggi in keluarga:
    print(tinggi)
```

```python
for index, tinggi in enumerate(keluarga):
    print("Index " + str(index) + ": " + str(tinggi))
```

```python
for huruf in 'keluarga':
    print(huruf.capitalize())
```

## Loop pada dictionary

```python
negara = { "afghanistan":30.55,
"albania":2.77,
"algeria":39.21 }
for key, value in negara.items():
    print(key + " -- " + str(value))
```

## NumPy Arrays

```python
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_berat / np_tinggi ** 2
for val in bmi:
    print(val)
```

```python
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
d_2 = np.array([np_tinggi, np_berat])
for val in d_2:
    print(val)
```

```python
import numpy as np
np_tinggi = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_berat = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
d_2 = np.array([np_tinggi, np_berat])
for val in np.nditer(d_2):
    print(val)
```

## Iterasi pada DataFrame

```python
import pandas as pd
negara = pd.read_csv('negara.csv', index_col=0)
negara
```

```python
for sesuatu in negara:
    print(sesuatu)
```

```python
for label, baris in negara.iterrows():
    print(label)
    print(baris)
```

```python
for label, baris in negara.iterrows():
    print(label + ": " + baris['capital'])
```

```python
negara
```

```python
for label, baris in negara.iterrows():
    negara.loc[label, "panjang_nama"] = len( baris["country"] )
negara
```
